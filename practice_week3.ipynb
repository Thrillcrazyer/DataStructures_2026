```python
import ctypes
import random
import time
from functools import wraps

def check_time(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs)
        end_time = time.time()
        print(f"{func.__name__} Execution time: {end_time - start_time:.6f} seconds")
        return result
    return wrapper
```

## 최대 부분 배열 문제

길이가 n 인 배열 $A$가 입력으로 주어 질 때, i번째부터 j 번째 까지 원소들로 이루어진 배열 $A[i],...A[j]$를 부분 배열이라고 하고 $A[i,j]$ 로 나타낸다.

이 때, 배열 $A$의 최대 부분 배열, 즉 부분 배열에 속한 원소들의 합이 최대가 되는 부분 배열을 찾고 그 부분 배열의 합을 구하는 알고리즘을 작성하시오.

- 예를 들어, +31, -41, **+59, +26, -53, +58, +97,** -93, -23, +84 와 같은 배열이 있을 때, 최대 부분 배열은 $A[2,6]$ 으로 합의 크기는 187 이다.

### 1. $O(n^3)$ 아이디어

$$T(n) = \sum_{i=0}^{n-1} \sum_{j=i}^{n-1} \sum_{k=i}^{j} 1 = \sum_{i=0}^{n-1} \sum_{j=i}^{n-1} (j - i + 1)$$

$$= \sum_{i=0}^{n-1} \frac{(n-i)(n-i+1)}{2} = \sum_{p=1}^{n} \frac{p(p+1)}{2}$$

$$= \frac{1}{2}\left(\frac{n(n+1)(2n+1)}{6} + \frac{n(n+1)}{2}\right) = \frac{n(n+1)(n+2)}{6}$$

$$\therefore \quad T(n) = \frac{n(n+1)(n+2)}{6} \in O(n^3)$$


```python
@check_time
def max_subarray_sum1(A: ctypes.Array, n: int) -> int:
    max_sum=0
    for i in range(n):
        for j in range(i, n):
            s=0
            for k in range(i, j+1):
                s=s+A[k]
            if s > max_sum:
                max_sum=s
                
    return max_sum

n=500
Array=(ctypes.c_int*n)()
for i in range(n):
    Array[i]=random.randint(-100,100)

max_sum1 = max_subarray_sum1(Array, n)
print(f"Max subarray sum (O(n^3)): {max_sum1}")
```

### 2. $O(n^2)$ 아이디어

안쪽 루프에서 $A[i..j]$의 합을 매번 처음부터 다시 계산하지 않고, 이전 합에 $A[j]$를 누적하여 중복 계산을 제거한다.

$$T(n) = \sum_{i=0}^{n-1} \sum_{j=i}^{n-1} 1 = \sum_{i=0}^{n-1} (n - i) = \sum_{p=1}^{n} p = \frac{n(n+1)}{2}$$

$$\therefore \quad T(n) = \frac{n(n+1)}{2} \in O(n^2)$$


```python
@check_time
def max_subarray_sum2(A: ctypes.Array, n: int) -> int:
    max_sum=0
    for i in range(n):
        this_sum = 0
        for j in range(i, n):
            this_sum = this_sum + A[j]
            if this_sum > max_sum:
                max_sum = this_sum
     
    return max_sum

n=500
Array=(ctypes.c_int*n)()
for i in range(n):
    Array[i]=random.randint(-100,100)

max_sum2 = max_subarray_sum2(Array, n)
print(f"Max subarray sum (O(n^2)): {max_sum2}")
```

### 3. $O(n)$ 아이디어 

배열을 한 번만 순회하면서, 현재 위치까지의 누적합이 음수가 되면 버리고 0부터 다시 시작한다.

$$\text{this\_sum}(i) = \max\!\big(\text{this\_sum}(i-1) + A[i],\; 0\big)$$

$$\text{max\_sum} = \max_{0 \le i \le n-1} \text{this\_sum}(i)$$

루프가 하나이므로:

$$T(n) = \sum_{i=0}^{n-1} 1 = n$$

$$\therefore \quad T(n) = n \in O(n)$$


```python
@check_time
def max_subarray_sum3(A: ctypes.Array, n: int) -> int:
    max_sum=0
    this_sum=0
    for i in range(n):
        this_sum = max(this_sum + A[i], 0)
        if this_sum > max_sum:
            max_sum = this_sum

    return max_sum

n=500
Array=(ctypes.c_int*n)()
for i in range(n):
    Array[i]=random.randint(-100,100)

max_sum3 = max_subarray_sum3(Array, n)
print(f"Max subarray sum (O(n)): {max_sum3}")
```
